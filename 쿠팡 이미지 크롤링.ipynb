```python
# 쿠팡 크롤링
from openpyxl import Workbook as wb
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pyperclip
from urllib.request import urlopen
from urllib.parse import quote_plus
from openpyxl import load_workbook

# 엑셀 파일 불러오기
load_wb = load_workbook("C:/Users/LENOVO/Desktop/QA자료/제품.xlsx", data_only=True)
load_ws = load_wb['Sheet1']

# 제품 번호 물러오기(행만 바뀜)
get_number = load_ws['A1':'A249']
#  브랜드명 불러오기(행만 바뀜)
get_Nbrand = load_ws['B1':'B249']
# 제품명 불러오기(행만 바뀜)
get_Nprod = load_ws['C1':'C249']
# 모델명 불러오기(행만 바뀜)
get_Nmodel = load_ws['D1':'D249']

driver = webdriver.Chrome('C:/Users/LENOVO/Desktop/QA자료/chromedriver')
url = "https://www.coupang.com/"
driver.get(url)

i = 0

while(1): # 검색 반복
    
    search = driver.find_element_by_xpath('//*[@id="headerSearchKeyword"]')
    
    search.click()
    search.clear()
    pyperclip.copy(get_Nbrand[i][0].value+" "+get_Nmodel[i][0].value)# 검색어 입력
    search.send_keys(Keys.CONTROL, 'v')
    time.sleep(1)
    search_btn = driver.find_element_by_xpath('//*[@id="headerSearchBtn"]')
    search_btn.click()
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    source=soup.select('li.search-product ')
    if len(source) <= 3:
        leng = len(source)
    else:
        leng = 4
    for j in range(1,leng):
        detail_btn = driver.find_element_by_xpath('//*[@id='+str(source[j]['id'])+']/a/dl/dt/img').click()
        time.sleep(1)
        driver.switch_to.window(driver.window_handles[1]) # 새로운 창으로 이동
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        img = "https:"+soup.select('#repImageContainer > img')[0]['src']
        with urlopen(img) as f:
            with open('C:/Users/LENOVO/Desktop/QA자료/크롤링사진/' +get_number[i][0].value+"_"+str(j)+'.jpg','wb') as h:
                img = f.read()
                h.write(img)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(1)
    i += 1
    if i == 250 :
        break
```
